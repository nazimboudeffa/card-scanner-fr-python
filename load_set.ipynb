```python
import json
import sqlite3

with open("reference_images/mega-evolution/fr-fr/mega_evolution.json", "r", encoding="utf-8") as f:
    cards = json.load(f)

conn = sqlite3.connect("database.db")
c = conn.cursor()
c.execute("CREATE TABLE IF NOT EXISTS cartes (numero INTEGER, set TEXT, nom TEXT, rarete TEXT)")

for card in cards:
    c.execute("INSERT INTO cartes VALUES (:numero, :set, :nom, :rarete)", card)

conn.commit()
conn.close()
```
